### Increasing the block size

```python
%%bash
rm /tmp/encoded*
python ../src/VQ.py encode -b 8
rm /tmp/decoded.png
python ../src/VQ.py decode -b 8
python ../src/RDE.py
```

```python
display(Image(filename="/tmp/decoded.png"))
```

### Decreasing the number of 2D vectors

```python
%%bash
rm /tmp/encoded*
python ../src/VQ.py encode -q 32
rm /tmp/decoded.png
python ../src/VQ.py decode -q 32
python ../src/RDE.py
```

```python
display(Image(filename="/tmp/decoded.png"))
```

### Using both parameters

```python
%%bash
rm /tmp/encoded*
python ../src/VQ.py encode -q 32 -b 8
rm /tmp/decoded.png
python ../src/VQ.py decode -q 32 -b 8
python ../src/RDE.py
```

```python
display(Image(filename="/tmp/decoded.png"))
```

### Using Huffman

```python
%%bash
rm /tmp/encoded*
python ../src/VQ.py encode -q 32 -b 8 -c Huffman
rm /tmp/decoded.png
python ../src/VQ.py decode -q 32 -b 8 -c Huffman
python ../src/RDE.py
```

### Using zlib

```python
%%bash
rm /tmp/encoded*
python ../src/VQ.py encode -q 32 -b 8 -c z_lib
rm /tmp/decoded.png
python ../src/VQ.py decode -q 32 -b 8 -c z_lib
python ../src/RDE.py
```

### Using PNM

```python
%%bash
rm /tmp/encoded*
python ../src/VQ.py encode -q 32 -b 8 -c PNM
rm /tmp/decoded.png
python ../src/VQ.py decode -q 32 -b 8 -c PNM
python ../src/RDE.py
```

### Blurring the output

```python
%%bash
rm /tmp/encoded*
python ../src/VQ.py encode -q 32 -b 8
rm /tmp/decoded.png
python ../src/VQ.py decode -q 32 -b 8 -f gaussian_blur -s 5
python ../src/RDE.py
```
