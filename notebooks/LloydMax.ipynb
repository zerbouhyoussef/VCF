### Increasing QSS

```python
%%bash
rm /tmp/encoded*
python ../src/LloydMax.py encode -q 64
rm /tmp/decoded.png
python ../src/LloydMax.py decode -q 64
python ../src/RDE.py
```

```python
display(Image(filename="/tmp/decoded.png"))
```

### Using PNM

```python
%%bash
rm /tmp/encoded*
python ../src/LloydMax.py encode -q 64 -c PNM
rm /tmp/decoded.png
python ../src/LloydMax.py decode -q 64 -c PNM
python ../src/RDE.py -c /tmp/encoded.pnm
```

```python
display(Image(filename="/tmp/decoded.png"))
```

### Using PNG

```python
%%bash
rm /tmp/encoded*
python ../src/LloydMax.py encode -q 64 -c PNG
rm decoded.png
python ../src/LloydMax.py decode -q 64 -c PNG
python ../src/RDE.py -c /tmp/encoded.png
```

```python
display(Image(filename="/tmp/decoded.png"))
```

### Using Huffman

```python
%%bash
rm /tmp/encoded*
python ../src/LloydMax.py encode -q 64 -c Huffman
rm decoded.png
python ../src/LloydMax.py decode -q 64 -c Huffman
python ../src/RDE.py -c /tmp/encoded.huf
```

```python
display(Image(filename="/tmp/decoded.png"))
```

### Using zlib

```python
%%bash
rm /tmp/encoded*
python ../src/LloydMax.py encode -q 64 -c z_lib
rm decoded.png
python ../src/LloydMax.py decode -q 64 -c z_lib
python ../src/RDE.py -c /tmp/encoded.npz
```

```python
display(Image(filename="/tmp/decoded.png"))
```

### Using a blurring filter

```python
%%bash
rm /tmp/encoded*
python ../src/LloydMax.py encode -q 64
rm decoded.png
python ../src/LloydMax.py decode -q 64 -f gaussian_blur -s 5
python ../src/RDE.py
```

```python
display(Image(filename="/tmp/decoded.png"))
```
