### Using LloydMax quantization

```python
%%bash
rm /tmp/encoded*
python ../src/YCoCg.py encode -a LloydMax -q 32 -m -128 -n 127
rm /tmp/decoded.png
python ../src/YCoCg.py decode -a LloydMax -q 32 -m -128 -n 127
python ../src/RDE.py
```

```python
display(Image(filename="/tmp/decoded.png"))
```

### Using (2D) VQ

```python
%%bash
rm /tmp/encoded*
python ../src/YCoCg.py encode -a VQ -q 32
rm /tmp/decoded.png
python ../src/YCoCg.py decode -a VQ -q 32
python ../src/RDE.py
```

```python
display(Image(filename="/tmp/decoded.png"))
```

### Using a blur filter

```python
%%bash
rm /tmp/encoded*
python ../src/YCoCg.py encode -a VQ -q 32
rm /tmp/decoded.png
python ../src/YCoCg.py decode -a VQ -q 32 -f gaussian_blur -s 5
python ../src/RDE.py
```

```python
display(Image(filename="/tmp/decoded.png"))
```
