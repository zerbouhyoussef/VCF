### Post-decompression filtering the output

```python
%%bash
pwd
rm -f /tmp/encoded*
python ../src/2D-DCT.py encode
rm /tmp/decoded.png
python ../src/2D-DCT.py decode -f gaussian_blur -s 11
python ../src/RDE.py
```

```python
display(Image(filename="/tmp/decoded.png"))
```

### Disable the reordering of the coefficients into subbands

```python
%%bash
rm /tmp/encoded*
python ../src/2D-DCT.py encode -x
rm /tmp/decoded.png
python ../src/2D-DCT.py decode -x
python ../src/RDE.py
```

```python
display(Image(filename="/tmp/decoded.png"))
```

### Enable perceptual quantization

```python
%%bash
rm /tmp/encoded*
python ../src/2D-DCT.py encode -p -q 4 -B 32
rm /tmp/decoded.png
python ../src/2D-DCT.py decode -p -q 4 -B 32
python ../src/RDE.py
```

```python
display(Image(filename="/tmp/decoded.png"))
```

###  The same, but compressing more

```python
%%bash
rm /tmp/encoded*
python ../src/2D-DCT.py encode -p -q 8 -B 32
rm /tmp/decoded.png
python ../src/2D-DCT.py decode -p -q 8 -B 32
python ../src/RDE.py
```

```python
display(Image(filename="/tmp/decoded.png"))
```

## Using LloydMax

```python
%%bash
rm /tmp/encoded*
python ../src/2D-DCT.py encode -a LloydMax -m -2048 -n 2047
rm /tmp/decoded.png
python ../src/2D-DCT.py decode -a LloydMax -m -2048 -n 2047
python ../src/RDE.py
```

```python
display(Image(filename="/tmp/decoded.png"))
```
